```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\HP\\OneDrive\\Documents\\Sleep_health_and_lifestyle_dataset.csv")
```

```python
print("Dataset shape:", df.shape)
print(df.head())
```

```python
display(df.info())
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
```

```python
df.columns
```

```python
df["Sleep_Duration"] = pd.to_numeric(df["Sleep_Duration"], errors="coerce")
```

```python
print('\nMissing values per column:')
print(df.isnull().sum())
```

```python
dups=df.duplicated().sum()
print("Duplicate count:", dups)
```

```python
print(df.describe())
```

```python
df["Sleep_Disorder"] = df["Sleep_Disorder"].fillna("None").str.strip().str.title()
df["Sleep_Disorder"].replace({"None": "No Disorder"}, inplace=True)
```

```python
num_cols = ["Age", "Sleep_Duration", "Quality_of_Sleep", 
            "Physical_Activity_Level", "Stress_Level", 
            "Heart_Rate", "Daily_Steps"]

sns.set(style="whitegrid", palette="muted")

for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True, bins=20, color="skyblue", edgecolor="black")
    plt.title(f"Distribution of {col}")
    plt.show()
cat_cols = ["Gender", "Occupation", "BMI_Category", "Sleep_Disorder"]

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, palette="pastel", edgecolor="black")
    plt.title(f"Count of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="Blues", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Gender", y="Sleep_Duration", palette="Set2")
plt.title("Sleep Duration by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="BMI_Category", y="Quality_of_Sleep", palette="Set3")
plt.title("Quality of Sleep by BMI Category")
plt.show()
```

```python

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Sleep_Disorder", y="Stress_Level", palette="Set1")
plt.title("Stress Level by Sleep Disorder")
plt.show()
```

```python
print("\n============================")
print(" SUMMARY INSIGHTS")
print("============================")
print("- Individuals with sleep disorders tend to have higher stress levels and shorter sleep durations.")
print("- Strong negative relationship between Stress Level and Sleep Duration indicates stress reduces sleep quality.")
print("- Physical activity appears to improve sleep quality moderately.")
print("- Most participants without disorders show lower stress and higher quality of sleep.")
print("- Gender differences are minor but may show slightly better sleep among females.")
print("- Age distribution suggests balanced representation without extreme outliers.")
print("- No extreme missing data or irregular values detected in most numeric fields.")
```
